```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

mort =pd.read_csv(r'C:\Users\Csucsu\Desktop\Infant mortality rate.csv',sep=';')
GDP = pd.read_csv(r'C:\Users\Csucsu\Desktop\GDP growth (annual %).csv',sep=';')
net = pd.read_csv(r'C:\Users\Csucsu\Desktop\net.txt', delimiter='|')
life = pd.read_csv(r'C:\Users\Csucsu\Desktop\Life expectancy.txt', sep='|')
```

```python
mort['Year(s)'].replace('(-.*)','',regex=True, inplace=True)
% matplotlib inline
fig1, ax = plt.subplots(figsize=(10,4))
y1 = mort[mort['Country or Area'] == 'Hungary']['Value'].tolist()
x = mort[mort['Country or Area'] == 'Hungary']['Year(s)'].tolist()
plt.plot(x, y1, color='#FF530D', linewidth=4.0, label='Hungary')
plt.plot(mort[mort['Country or Area'] == 'United States of America']['Year(s)'], 
         mort[mort['Country or Area'] == 'United States of America']['Value'], color='#2E10A7', linewidth=4.0, label='USA')
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('Years', alpha=0.6)
ax.set_ylabel('Infant deaths / 1000 births', alpha=0.6)
plt.title('Infant Mortality rate', alpha=0.7)
plt.xticks(alpha=0.6)
plt.yticks(alpha=0.6)
ax.legend(bbox_to_anchor=(0.85, 0.8),
           bbox_transform=plt.gcf().transFigure)
ax.tick_params(right=False, top=False)
legend = ax.legend()
legend.draw_frame(False)
plt.show()
```
